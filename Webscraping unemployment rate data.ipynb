```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_site = "https://www.macrotrends.net/countries/AFG/afghanistan/unemployment-rate"
```

```python
response = requests.get(base_site)
```

```python
response
```

```python
html = response.content
html[:100]
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
with open('afghanistan_unemployment_rate.html','wb') as file:
    file.write(soup.prettify('utf-8'))
```

### searching and navigating the HTML tree using find and find_all

```python
soup.find('head')
```

```python
# video tag
print(soup.find('video'))
```

```python
# lets see if there is any anchor tags (or a) marks the links
soup.find('a')

# returns the first result
```

```python
links = soup.find_all('a')

# returns all the list of all links
isinstance(links,list) 
```

```python
len(links)
```

```python
# lets store a body of the table, 

table = soup.find('tbody')

```
